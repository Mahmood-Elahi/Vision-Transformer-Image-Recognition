```python
import os
import json
import matplotlib.pyplot as plt
from PIL import Image
```

## Final Model Summary

The best model used:

- Dataset: CIFAR-10
- Model: Pure Vision Transformer
- Patch size: 2 × 2
- Tokens: 256 image tokens + 1 CLS token
- Embedding dimension: 384
- Depth: 12 transformer blocks
- Attention heads: 6
- Best validation accuracy: 90.98%
- Final test accuracy: 91.06%
- Training GPU: NVIDIA RTX A6000

```python
cifar10_classes = [
    "airplane",
    "automobile",
    "bird",
    "cat",
    "deer",
    "dog",
    "frog",
    "horse",
    "ship",
    "truck",
]

cifar10_classes
```

```python
graph_paths = {
    "Validation Accuracy": "../results/graphs/validation_accuracy_pure_vit.png",
    "Training and Validation Loss": "../results/graphs/loss_curve_pure_vit.png",
    "Learning Rate Schedule": "../results/graphs/learning_rate_pure_vit.png",
    "Confusion Matrix": "../results/graphs/confusion_matrix_pure_vit.png",
}

for title, path in graph_paths.items():
    if os.path.exists(path):
        img = Image.open(path)
        plt.figure(figsize=(8, 6))
        plt.imshow(img)
        plt.axis("off")
        plt.title(title)
        plt.show()
    else:
        print(f"Missing file: {path}")
```

## Performance Analysis

The model performs best on visually distinct classes such as:

- automobile
- ship
- truck
- frog

The most difficult classes are:

- cat
- dog
- bird

This is expected because CIFAR-10 images are only 32 × 32 pixels. Animal classes can look similar at low resolution.
